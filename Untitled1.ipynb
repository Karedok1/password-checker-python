```python
import re
import hashlib

# mengecek kekuatan password:

def cek_password(password):
    if len(password) < 8:
        return "❌ password lemahc(minimal 8 karakter)"
    if not re.search("[A-Z]", password):
        return "❌ harus mengandung huruf besar"
    if not re.search("[a-z]", password):
        return "❌ harus mengandung huruf kecil"
    if not re.search("[0-9]", password):
        return "❌ harus mengandung angka"
    if not re.search("[!@#$%^&*]", password):
        return "❌ harus mengandung [!@#$%..etc]"

    return " ✅ password kuat!.."


def daftar_akun():
    print("selamat datang !")
    name = input("masukan nama: ")

    while True:
        password = input("password: ")
        hasil = cek_password(password)
        print(hasil)
        if "✅" in hasil:
            break



    hash_pw = hashlib.sha256(password.encode()).hexdigest()

    #simpan file

    with open("akun.txt", "a") as f:
        f.write(f"{name}, {hash_pw}\n")

    print("akun berhasil dibuat✅")

daftar_akun()

```
